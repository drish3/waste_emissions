```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
oper_co2 = pd.read_csv('oper_co2.csv')
plt.style.use('tableau-colorblind10')
ax = oper_co2.plot.area(color=('#595959','#ababab','#2ca02c','#ff800e','#a2c8ec'))
plt.rc('axes')
plt.xlim([0,28])
fontsize = 12
plt.title("Annual CO2 emissions from fuel shipping", fontsize = 16)
plt.xlabel("Year", fontsize = 14)
plt.ylabel("kg", fontsize = 14)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', prop={'size': 12})
plt.xticks(fontsize=12 )
plt.yticks(fontsize=12 )
xticks = [0, 3, 8, 13, 18, 23, 28]
xtick_labels = [2022, 2025, 2030, 2035, 2040, 2045, 2050]
plt.xticks(xticks)
ax.set_xticklabels(xtick_labels)
plt.savefig("co2 emissions.jpg",bbox_inches='tight', dpi=900)
plt.show()
```
